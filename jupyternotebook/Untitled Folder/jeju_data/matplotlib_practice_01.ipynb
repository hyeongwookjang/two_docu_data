```python
plt.plot(x,y,'b--o',label='Data 1')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')
plt.legend()
```

```python
hf = plt.figure()
ha = hf.add_axes([0,0,2,2])
ha.plot(x,y)
```

```python
fig, ax = plt.subplots()
line, = ax.plot(x,y)
```

```python
x_new = np.array([10,11,12])
y_new = np.array([10,10,10])
line.set_data(x_new,y_new)
fig
```

```python
i = 1.8
x_new = np.array([10,11,12])
y_new = np.array([10,10,10])
line.set_data(x_new,i*y_new)
fig
```
