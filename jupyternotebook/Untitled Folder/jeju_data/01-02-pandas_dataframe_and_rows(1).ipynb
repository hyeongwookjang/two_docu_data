```python
import pandas as pd
```

# 파이썬 판다스 라이브러리 연습

```python
df = pd.DataFrame(
        {"a" : [4 ,5, 6],
        "b" : [7, 8, 9],
        "c" : [10, 11, 12]},
        index = [1, 2, 3])
df
# 함수가 그 형식에 맞게 기능을 수행하는 역할
```

```python
df = pd.DataFrame(
        [[4, 7, 10],
        [5, 8, 11],
        [6, 9, 12]],
        index=[1, 2, 3],
        columns=['a', 'b', 'c'])
df
```

```python
df
```

```python
df.loc[3, "a"]
```

```python
df.loc[[1, 2], ["a", "b"]]
```

```python
df = pd.DataFrame(
        [[4, 7, 10],
        [5, 8, 11],
        [6, 9, 12]],
        index=[1, 2, 3],
        columns=['a', 'b', 'c'])
# 리스트형태의 데이터를 판다스에선 시리즈라고 부른다.
# 시리즈들도 데이터프레임에 들어가면 변환된다.
df
```

```python
df["a"]
# df[["a","b"]]
```

```python
df.loc[3, "a"]
# location으로 좌표값 호출
```

```python
df.loc[[1, 2], ["a", "b"]]
# 리스트로 감싸면 여러개 가능
```

```python
import numpy as np
df = pd.DataFrame(
        {"a" : [4 ,5, 6, 6, np.nan],
        "b" : [7, 8, np.nan, 9, 9],
        "c" : [10, 11, 12, np.nan, 12]},
        index = pd.MultiIndex.from_tuples(
        [('d',1),('d',2),('e',2),('e',3),('e',4)],
        names=['n','v']))
df
```
