```python
j_age = age_data[age_data['행정구역'].str.contains('5011000000')]
s_age = age_data[age_data['행정구역'].str.contains('5013000000')]
```

```python
# 0~9, 10~19, 10살 단위로 묶어서 비교

def set_age(data):
  k = 3
  for i in range(10):
    컬럼명 = str(i * 10) + '~' + str(i * 10 + 9) + '세 합계' # 컬럼명 지정
    데이터 = data.iloc[0:1, k:k+10]
    데이터 = 데이터.apply(pd.to_numeric).sum(axis=1) # 숫자로 변환
    data.loc[:, 컬럼명] = 데이터
    k += 10

set_age(j_age)
set_age(s_age)
```
