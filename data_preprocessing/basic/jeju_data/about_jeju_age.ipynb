```python
# 각 시별 데이터합계가 이미 있기 때문에 이 데이터 사용
j_age = age_data[age_data['행정구역'].str.contains('5011000000')]
s_age = age_data[age_data['행정구역'].str.contains('5013000000')]
```

```python
# 이번엔 0~9, 10~19살 10살단위로 묶어서 비교
def what_age(data):
  원본컬럼 = 3
  for i in range(10):
    컬럼명 = str(i*10) + '~' + str(i*10+9)+'세 합계' # 컬럼명 지정
    데이터 = data.iloc[0:1, 원본컬럼:원본컬럼+10]
    데이터 = 데이터.apply(pd.to_numeric).sum(axis=1) # 숫자로 변환 -> 합
    data.loc[:, 컬럼명] = 데이터
    원본컬럼 += 10

what_age(j_age)
what_age(s_age)
```
